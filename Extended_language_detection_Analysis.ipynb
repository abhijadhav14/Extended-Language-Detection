```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load the data
file_path = '/content/drive/My Drive/Dataset/extended_language_detection_data_set_3000.xlsx'
data = pd.read_excel(file_path)
```

```python
# Count total unique languages
unique_languages = data['Language'].nunique()
print(f"Total unique languages: {unique_languages}")

# Statistics for word count
word_count_stats = data['Word Count'].describe()
print("Word Count Statistics:\n", word_count_stats)
```

```python
# Plot 1: Language distribution
plt.figure(figsize=(12, 6))
sns.countplot(data=data, y='Language', order=data['Language'].value_counts().index)
plt.title('Language Distribution')
print("Language Distribution:\n", data['Language'].value_counts())
plt.xlabel('Count')
plt.ylabel('Language')
plt.show()
```

```python
# Plot 2: Average Word Length vs. Language Family
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Language Family', y='Average Word Length')
plt.title('Average Word Length by Language Family')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 3: Special Characters vs. Language
plt.figure(figsize=(12, 6))
sns.barplot(data=data, x='Language', y='Special Characters', order=data.groupby('Language')['Special Characters'].mean().sort_values(ascending=False).index)
plt.title('Average Special Characters by Language')
plt.xlabel('Language')
plt.ylabel('Average Special Characters')
plt.xticks(rotation=45)
plt.show()
```
